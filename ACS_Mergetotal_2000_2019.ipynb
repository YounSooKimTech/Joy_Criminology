```python
import pandas as pd

df1 = pd.read_csv("https://raw.githubusercontent.com/YounSooKimTech/Joy_Criminology/main/Data/ACS_2000_2004.csv")
df2 = pd.read_csv("https://raw.githubusercontent.com/YounSooKimTech/Joy_Criminology/main/Data/ACS_2005_2009.csv")
df3 = pd.read_csv("https://raw.githubusercontent.com/YounSooKimTech/Joy_Criminology/main/Data/ACS_2010_2022.csv")
```

```python
df_acs = pd.concat([df1, df2, df3])

df_acs.drop("Geographic Name", axis=1, inplace=True)

df_acs.head()
```

```python
df_acs["Year"].value_counts().sort_index()
```

```python
df_acs_yr = df_acs[(df_acs["Year"]>=2000) & (df_acs["Year"]<=2019)]

df_acs_yr.to_csv("State_Level_ACS.csv", index=False)
```

```python
import plotly.express as px
import kaleido

fig = px.choropleth(df_acs_yr,
                    locations="State",
                    locationmode='USA-states',
                    color="Percent White",
                    range_color = [25,88],
                    color_continuous_scale="Blues",
                    scope='usa',
                    animation_frame='Year',
                    title = "Percent White")
fig.update_layout(
    margin=dict(l=0, r=0, t=50, b=0),  # Adjust the left, right, top, and bottom margins
    paper_bgcolor="white",  # Set the background color
    geo=dict(
        lakecolor="rgb(255, 255, 255)",  # Set the color of lakes
    )
)

fig.write_image("Whie_Proportion.png")
fig.show()
```
