```python
import pandas as pd

```

```python
df = pd.read_csv("/content/Profiles0402000.csv")

df_temp = df[ ((df["Table Number"] == 1) & (df["Order"].isin([1.0, 2, 17, 25]))) |
    ((df["Table Number"] == 2) & (df["Order"] == 16)) |
    ((df["Table Number"] == 3) & (df["Order"].isin([6, 89]))) ]

df_temp = df_temp.drop(["Geographic ID Code", "Table Number", "Order", "Lower Bound", "Upper Bound"], axis=1)

df_temp_long = df_temp.pivot(index='Geographic Name', columns='Stub', values='Estimate').reset_index()
df_temp_long.columns = df_temp_long.columns.str.replace('\n', '').str.replace('[()]', '', regex=True)
df_temp_long.columns = df_temp_long.columns.str.replace("bachelor's degree", "BA")
df_temp_long

df_temp_long.iloc[:, 1:] = df_temp_long.iloc[:, 1:].replace("," , "", regex=True)
df_temp_long.iloc[:, 1:] = df_temp_long.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')


df_temp_long["Year"] = 2000
df_temp_long["Percent Male"] = round(df_temp_long["Male"]/df_temp_long["Total population"] * 100, 2)
df_temp_long["Percent White"] = round(df_temp_long["White"]/df_temp_long["Total population"] * 100, 2)

df_2000 = df_temp_long[["Geographic Name", "Year", "Median age years", "Percent Male", "Percent White",
              "Percent BA or higher", "Percent unemployed", "Per capita income dollars" ]]

#df_2000
```

```python
df = pd.read_csv("/content/Profiles0402001.csv")

df_temp = df[( (df["Table"] == 1) & df["Subtable"].isin([1.0, 2, 17, 25]) |
    (df["Table"] == 2) & (df["Subtable"] == 16) |
    (df["Table"] == 3) & df["Subtable"].isin([6, 91]) )]

df_temp = df_temp.drop(["Table", "Subtable", "Unnamed: 5","Unnamed: 6"], axis=1)

df_temp_long = df_temp.pivot(index='Geographic Name', columns='Stub', values='Estimate').reset_index()

df_temp_long.columns = df_temp_long.columns.str.replace('\n', '').str.replace('[()]', '', regex=True)
df_temp_long.columns = df_temp_long.columns.str.replace("bachelor's degree", "BA")

df_temp_long.iloc[:, 1:] = df_temp_long.iloc[:, 1:].replace("," , "", regex=True)
df_temp_long.iloc[:, 1:] = df_temp_long.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

df_temp_long["Year"] = 2001
df_temp_long["Percent Male"] = round(df_temp_long["Male"]/df_temp_long["Total population"] * 100, 2)
df_temp_long["Percent White"] = round(df_temp_long["White"]/df_temp_long["Total population"] * 100, 2)


df_2001 = df_temp_long[["Geographic Name", "Year", "Median age years", "Percent Male", "Percent White",
              "Percent BA or higher", "Percent unemployed", "Per capita income dollars" ]]
#df_2001
```

```python
df = pd.read_csv("/content/Profiles0402002.csv")
df_temp = df [(df["Table"]==1) & df["Table Order"].isin([2, 4, 20, 31]) |
    (df["Table"] == 2) & (df["Table Order"] == 21) |
    (df["Table"] == 3) & df["Table Order"].isin([8, 111]) ]

df_temp = df_temp[["Geographic Name", "Stub", "Estimate"]]


df_temp_long = df_temp.pivot(index='Geographic Name', columns='Stub', values='Estimate').reset_index()
df_temp_long.columns = df_temp_long.columns.str.replace('\n', '').str.replace('[()]', '', regex=True)
df_temp_long.columns = df_temp_long.columns.str.replace("bachelor's degree", "BA")

df_temp_long.iloc[:, 1:] = df_temp_long.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

df_temp_long["Year"] = 2002
df_temp_long["Percent Male"] = round(df_temp_long["Male"]/df_temp_long["Total population"] * 100, 2)
df_temp_long["Percent White"] = round(df_temp_long["White"]/df_temp_long["Total population"] * 100, 2)

df_2002 = df_temp_long[["Geographic Name", "Year", "Median age years", "Percent Male", "Percent White",
              "Percent BA or higher", "Percent unemployed", "Per capita income dollars" ]]
#df_2002
```

```python
df = pd.read_csv("/content/Profiles0402003.csv")


df[(df["Table Number"] == 1) & (df["Table Order"].isin([2, 4, 20, 31]))]
df[(df["Table Number"] == 2) & (df["Table Order"] == 21)]
df[(df["Table Number"] == 3) & (df["Table Order"].isin([8, 109]))]

df_temp = df [ (df["Table Number"] == 1) & (df["Table Order"].isin([2, 4, 20, 31])) |
     (df["Table Number"] == 2) & (df["Table Order"] == 21) |
    (df["Table Number"] == 3) & df["Table Order"].isin([8, 109]) ]


df_temp = df_temp[["Geographic Name", "Stub", "Estimate"]]

df_temp_long = df_temp.pivot(index='Geographic Name', columns='Stub', values='Estimate').reset_index()
df_temp_long.columns = df_temp_long.columns.str.replace('\n', '').str.replace('[()]', '', regex=True)
df_temp_long.columns = df_temp_long.columns.str.replace("bachelor's degree", "BA")

df_temp_long.iloc[:, 1:] = df_temp_long.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

df_temp_long["Year"] = 2003
df_temp_long["Percent Male"] = round(df_temp_long["Male"]/df_temp_long["Total population"] * 100, 2)
df_temp_long["Percent White"] = round(df_temp_long["White"]/df_temp_long["Total population"] * 100, 2)

df_2003 = df_temp_long[["Geographic Name", "Year", "Median age years", "Percent Male", "Percent White",
              "Percent BA or higher", "Percent unemployed", "Per capita income dollars" ]]

```

```python
df = pd.read_csv("/content/Profiles0402004.csv")

df[(df["Table Number"] == 1) & (df["Table Order"].isin([1, 3, 19, 40]))]
df[(df["Table Number"] == 2) & (df["Table Order"] == 20)]
df[(df["Table Number"] == 3) & (df["Table Order"].isin([11, 111]))]

df_temp = df [ (df["Table Number"] == 1) & (df["Table Order"].isin([1, 3, 19, 40])) |
     (df["Table Number"] == 2) & (df["Table Order"] == 20) |
    (df["Table Number"] == 3) & df["Table Order"].isin([11, 111]) ]

df_temp = df_temp[["Geographic Name", "Stub", "Estimate"]]

df_temp_long = df_temp.pivot(index='Geographic Name', columns='Stub', values='Estimate').reset_index()
df_temp_long.columns = df_temp_long.columns.str.replace('\n', '').str.replace('[()]', '', regex=True)
df_temp_long.columns = df_temp_long.columns.str.replace("bachelor's degree", "BA")

df_temp_long.iloc[:, 1:] = df_temp_long.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

df_temp_long["Year"] = 2004
df_temp_long["Percent Male"] = round(df_temp_long["Male"]/df_temp_long["Total population"] * 100, 2)
df_temp_long["Percent White"] = round(df_temp_long["White"]/df_temp_long["Total population"] * 100, 2)

df_temp_long.rename(columns={"Unemployed" : "Percent unemployed"}, inplace=True)


df_2004 = df_temp_long[["Geographic Name", "Year", "Median age years", "Percent Male", "Percent White",
              "Percent BA or higher", "Percent unemployed", "Per capita income dollars" ]]

# df_2004
```

```python
df_2000.columns
df_2001.columns
df_2002.columns
df_2003.columns
df_2004.columns

df_0004 = pd.concat([df_2000, df_2001, df_2002, df_2003, df_2004], ignore_index=True)

df_0004["Year"].value_counts()
```

```python
df_0004.columns
```

```python
! pip install us
import us

def state_name_to_abbreviation(state_name):
    state = us.states.lookup(state_name)
    if state:
        return state.abbr
    else:
        return state_name

df_0004['State'] = df_0004["Geographic Name"].apply(state_name_to_abbreviation)

df_0004["State"] = df_0004["State"].str.replace("District of Columbia", "DC")
```

```python
new_col_order = [df_0004.columns.tolist()[-1]] + df_0004.columns.tolist()[0:]

df_0004 = df_0004[new_col_order]
df_0004 = df_0004.iloc[:, :-1]
df_0004
```
