# Hate Crime 2000 ~ 2019

```python
import pandas as pd

df = pd.read_csv("/content/FBI_HateCrime.csv")

df.head()

df_hate = df[df["data_year"]>=2000]
df_hate = df_hate[df_hate["data_year"]<=2019]

df_hate_state = df_hate.groupby(["state_abbr", "data_year"])["incident_id"].count().reset_index()
```

```python
df_hate_state.columns= ["State", "Year", "Reported_Hate_Crime"]
df_hate_state
```

```python
temp = df_hate_state.groupby("State").mean().reset_index()
temp.columns = ["State", "Year", "Mean Number of Hate Crimes"]
```

```python
temp["Mean Number of Hate Crimes"].describe()
```

```python
import plotly.express as px
import kaleido

fig = px.choropleth(temp,
                    locations="State",
                    locationmode='USA-states',
                    color="Mean Number of Hate Crimes",
                    range_color=[25, 140],
                    color_continuous_scale="Blues",
                    scope='usa',
                    # animation_frame='Year',
                    title="Average Reported Hate Crime Across 2000 - 2022")

fig.update_layout(
    margin=dict(l=0, r=0, t=50, b=0),  # Adjust the left, right, top, and bottom margins
    paper_bgcolor="white",  # Set the background color
    geo=dict(
        lakecolor="rgb(255, 255, 255)",  # Set the color of lakes
    )
)


fig.write_image("Hate_Crime_2000_2019.png")
fig.show()
```

# Crime Data

```python
df = pd.read_csv("https://raw.githubusercontent.com/YounSooKimTech/Joy_Criminology/main/Data/FBI_Crime_total.csv")

df = df[["Area", "Year", "Violent_crime", "Property_crime", "Total_crime"]]

df
```

```python
! pip install us
import us

def state_name_to_abbreviation(state_name):
    state = us.states.lookup(state_name)
    if state:
        return state.abbr
    else:
        return state_name

df['State'] = df["Area"].apply(state_name_to_abbreviation)

df["State"] = df["State"].str.replace("District Of Columbia", "DC")

#df["State"].value_counts()
```

```python
df["Year"].value_counts().sort_index()
```

# Total crime

```python
temp_violent = df.groupby("State")["Total_crime"].mean().reset_index()
print(temp_violent["Total_crime"].describe())

temp_violent.columns = ["State", "Average Total Crime"]
```

```python
import plotly.express as px
import kaleido

fig = px.choropleth(temp_violent,
                    locations="State",
                    locationmode='USA-states',
                    color="Average Total Crime",
                    range_color=[52000, 240000],
                    color_continuous_scale="Darkmint",
                    scope='usa',
                    # animation_frame='Year',
                    title="Average Number Total Crime Across 2000 ~ 2019")

fig.update_layout(
    margin=dict(l=0, r=0, t=50, b=0),  # Adjust the left, right, top, and bottom margins
    paper_bgcolor="white",  # Set the background color
    geo=dict(
        lakecolor="rgb(255, 255, 255)",  # Set the color of lakes
    )
)

fig.write_image("Total_Crime_2000_2019.png")
fig.show()
```

# Sun Exposure

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/YounSooKimTech/Joy_Criminology/main/Data/Sun_Exposure.csv")

df = df[(df["Year"] >= 2000) & (df["Year"] <=2019)]

temp_sun = df.groupby("State")["UV_Irradiance"].mean().reset_index()
temp_sun.columns = ["State", "UV Irradiance (kj/m2)"]
temp_sun.describe()
```

```python
import plotly.express as px
import kaleido

fig = px.choropleth(temp_sun,
                    locations="State",
                    locationmode='USA-states',
                    color="UV Irradiance (kj/m2)",
                    range_color=[2500, 3400],
                    color_continuous_scale="Pinkyl",
                    scope='usa',
                    # animation_frame='Year',
                    title="Sun Exposure Measured by Solar Irradiance Across 2000 ~ 2019")

fig.update_layout(
    margin=dict(l=0, r=0, t=50, b=0),  # Adjust the left, right, top, and bottom margins
    paper_bgcolor="white",  # Set the background color
    geo=dict(
        lakecolor="rgb(255, 255, 255)",  # Set the color of lakes
    )
)

fig.write_image("Sun_Exposure_2000_2019.png")
fig.show()
```

# EPA

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/YounSooKimTech/Joy_Criminology/main/Data/EPA_AirPollutants_Yearly.csv")
df = df[(df["Year"] >= 2000) & (df["Year"] <=2019)]

df_PM = df[df["Pollutant"] == "PM2.5"]
df_PM.dropna(subset=["Pollutant"])

temp_PM = df_PM.groupby("State")["Observed_Value"].mean().reset_index()
temp_PM.columns = ["State", "PM2.5"]
temp_PM["PM2.5"].describe()
```

```python
df_PM
```

```python
import plotly.express as px
import kaleido

fig = px.choropleth(temp_PM,
                    locations="State",
                    locationmode='USA-states',
                    color="PM2.5",
                    range_color=[16, 19],
                    color_continuous_scale="Brwnyl",
                    scope='usa',
                    # animation_frame='Year',
                    title="Air Quality (PM2.5) Across 2000 ~ 2019")

fig.update_layout(
    margin=dict(l=0, r=0, t=50, b=0),  # Adjust the left, right, top, and bottom margins
    paper_bgcolor="white",  # Set the background color
    geo=dict(
        lakecolor="rgb(255, 255, 255)",  # Set the color of lakes
    )
)

fig.write_image("Air_PM2.5_2000_2019.png")
fig.show()
```
