```python
import requests
import pandas as pd

url = "https://api.census.gov/data/2005/acs/acs1/profile"

params = {
    'get': 'NAME,DP01_0001E,DP01_0002E,DP01_0017E,DP01_0032E,DP02_0016E,DP03_0087E,DP03_0009E',
    'for': 'state:*',
    'key': 'e998f014fb191e980df68677cc7d5d10ef61b516'
}

response = requests.get(url, params=params)

data = response.json()
columns = data[0]
rows = data[1:]

df = pd.DataFrame(rows, columns=columns)
df["Year"] = 2005

df.columns = ["State_full", "Total Population", "Male", "Median age","White", "Percent BA or More", "Per Capita Income", "Percent Unemployment", "State_n", "Year"]

df.drop("State_n", axis=1, inplace=True)
df_2005 = df
```

```python
import requests
import pandas as pd

url = "https://api.census.gov/data/2009/acs/acs1/profile"

params = {
    'get': 'NAME,DP05_0002PE,DP05_0017E,DP05_0032PE,DP02_0067E,DP03_0089E,DP03_0005PE',
    'for': 'state:*',
    'key': 'e998f014fb191e980df68677cc7d5d10ef61b516'
}

response = requests.get(url, params=params)

data = response.json()
columns = data[0]
rows = data[1:]

df = pd.DataFrame(rows, columns=columns)

df["Year"] = 2009
df.columns = ["State_full", "Percent Male", "Median age","Percent White", "Percent BA or More", "Per Capita Income", "Percent Unemployment", "State_n", "Year"]
df.drop("State_n", axis=1, inplace=True)

df_2009 = df
```

```python
df_0506 = pd.concat([df_2005, df_2006], ignore_index=True)
df_0506

df_0506.iloc[:, 1:] = df_0506.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

df_0506["Percent Male"] = (df_0506["Male"] / df_0506["Total Population"]) * 100
df_0506["Percent Male"] = round(df_0506["Percent Male"], 2)
df_0506["Percent White"] = (df_0506["White"] / df_0506["Total Population"]) * 100
df_0506["Percent White"] = round(df_0506["Percent White"], 2)

df_0506_temp = df_0506[['State_full', 'Percent Male', 'Median age', 'Percent White',
       'Percent BA or More', 'Per Capita Income', 'Percent Unemployment',
       'Year']]

```

```python
import pandas as pd

df_total = pd.concat([df_0506_temp, df_2007, df_2008, df_2009])

df_total.rename(columns = {"State_full":"Geographic Name"}, inplace=True)
df_total.rename(columns = {"Median age":"Median age years"}, inplace=True)
df_total.rename(columns = {"Per Capita Income":"Per capita income dollars"}, inplace=True)
df_total.rename(columns = {"Percent BA or More":"Percent BA or higher"}, inplace=True)
df_total.rename(columns = {"Percent Unemployment":"Percent unemployed"}, inplace=True)

df_total
```

```python
! pip install us
import us

def state_name_to_abbreviation(state_name):
    state = us.states.lookup(state_name)
    if state:
        return state.abbr
    else:
        return state_name

df_total['State'] = df_total["Geographic Name"].apply(state_name_to_abbreviation)

df_total["State"] = df_total["State"].str.replace("District of Columbia", "DC")

df_total["State"].value_counts()
```

```python
df_total = df_total[["State","Geographic Name","Year","Median age years","Percent Male",	"Percent White",	"Percent BA or higher","Percent unemployed","Per capita income dollars"]]
```

```python
df_total.to_csv("ACS_2005_2009.csv", index=False)
```
