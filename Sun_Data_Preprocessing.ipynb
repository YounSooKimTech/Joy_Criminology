```python
import pandas as pd

df = pd.read_csv("/content/Historical_Sunlight.csv")
df =df.iloc[:, 1:]

df.rename(columns = {"State":"State_full",
           "Value":"UV_Irradiance"}, inplace=True)

df
```

```python
import us

def state_name_to_abbreviation(state_name):
    state = us.states.lookup(state_name)
    if state:
        return state.abbr
    else:
        return state_name


df['State'] = df['State_full'].apply(state_name_to_abbreviation)

df["State"] = df["State"].str.replace("District of Columbia", "DC")
```

```python
df["UV_Irradiance"].describe()
```

```python
temp = df[["State", "Year", "UV_Irradiance"]]

temp.to_csv("Sun_Exposure.csv", index=False)
```

```python
import plotly.express as px
import kaleido

fig = px.choropleth(df,
                    locations="State",
                    locationmode='USA-states',
                    color="UV_Irradiance",
                    range_color = [1000,5000],
                    color_continuous_scale="Jet",
                    scope='usa',
                    animation_frame='Year',
                    title = "Annual Average Daily Dose of UV Irradiance (J/M2) per Year")
fig.show()

fig.write_image("Sun_UV_Irradiance_Yearly_By_State.svg")
fig.write_html("Sun_UV_Irradiance_Yearly_By_State.html")
```
